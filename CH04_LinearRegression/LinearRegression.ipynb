```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('..\data\CH04_Regression\ex0.txt',sep='\\t',engine='python',header=None,names=['x0','x1','y'])
data.head()
```

**矩阵求导速查表** ：
https://yq.aliyun.com/articles/411340

```python
x = data.loc[:,['x0','x1']]
y = data['y']
```

```python
# x.T * y
# x.T*x
xMat = np.mat(x.to_numpy())
yMat = np.mat(y.to_numpy()).T
```

```python
xTy = xMat.T*yMat
xTx = xMat.T * xMat
wMat = xTx.I * xTy
wMat
```

```python
x.min()
# 0.995731  0.014855
xx = np.ones(shape=(10000,2))
xx[:,1] = np.linspace(0,1,10000)
yy = xx * wMat
fig, ax = plt.subplots()
ax.plot(xx[:,1],yy, color='red')
ax.scatter(x['x1'],y)
```

```python
yHat = xMat * wMat

np.corrcoef(yHat.T, yMat.T)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,y)
reg.coef_
```
